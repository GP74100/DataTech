## Chargement, profilage et préparation du fichier Job_profile

```python
# J'importe le module pandas afin de créer des dataframe
import pandas as pd
```

```python
# J'assigne le chemin à une variable pour une question de lisibilité et de confidentialité
chemin_Job_profile = "D:\Cours Lepont\Bloc 2\Cours 150524 Capstone 2\Dataset\Job_profile.csv"
```

```python
# Je charge le fichier Job_profile 
df_Job_profile = pd.read_csv(chemin_Job_profile)
```

```python
# J'affiche le fichier Job_profile sous forme de DataFrame
df_Job_profile
```

```python
# On ne remarque rien de particulier excepté les valeurs manquantes
```

```python
# J'affiche toutes les colonnes du Dataframe
df_Job_profile.columns
```

```python
# Je consulte le type des données dans les colonnes
df_Job_profile.dtypes
```

```python
# On remarque que "Compensation" est décalé par rapport aux autres intitulés  , par conséquent on va supprimé les espaces
df_Job_profile = df_Job_profile.rename(columns={' Compensation ': 'Compensation'})
```

```python
# On vérifie
df_Job_profile.dtypes
```

```python
# Je consulte le nombre de valeurs manquantes dans le DataFrame par ordre décroissant
(df_Job_profile.isnull().sum() / df_Job_profile.shape[0]).sort_values(ascending=False)
```

```python
# J'effectue un strip afin de supprimer les espaces dans la colonne Compensation
df_Job_profile["Compensation"] = df_Job_profile["Compensation"].str.strip()
```

```python
# On controle
df_Job_profile
```

```python
# Je trie le dataframe par ordre croissant par l'EmployeeID
df_Job_profile.sort_values('EmployeeID', inplace=True)
```

```python
df_Job_profile
```

```python
# Je change l'intitulé de la colonne Bonus % par Bonus_% afin d'avoir des données concordantes et j'harmonise l'intitulé de la colonne
df_Job_profile = df_Job_profile.rename(columns={'Bonus %': 'Bonus_%'})
```

```python
df_Job_profile
```

```python
# J'exporte sous le format csv
df_Job_profile.to_csv('Job_profile_cleaned.csv',index=False)
```
