```python
# 2eme methode avec l'aide de copilot
def convert_date(date):
    formats_source = ['%b %d-%Y', "%m%d%Y", "%m/%d/%Y", '%B %d, %Y']
    format_cible = '%d/%m/%Y'
    for format_source in formats_source:
        try:
            date = pd.to_datetime(date, format=format_source)
            date = date.strftime(format_cible)
        except ValueError:
            continue
    return date

df_companydata['Start_Date']= (df_companydata['Start_Date']).apply(convert_date)
df_companydata['Termination_Date']= (df_companydata['Termination_Date']).apply(convert_date)
df_companydata['Termination_Date'] = pd.to_datetime(df_companydata['Termination_Date'], dayfirst=True)
df_companydata['Start_Date'] = pd.to_datetime(df_companydata['Start_Date'], dayfirst=True)


```

```python
# En affichant le dataframe, on s'aperçoit que John Simpson index 104964 a terminé son contrat le 27/11/2022 mais dans la colonne 'active status', 
# il est indiqué 1 donc actif, il s'agit d'une erreur qui peut être touche d'autres employés.
 
# Nous allons donc recréer une colonne 'active status' corrigée qui prendra en compte la date de fin de contrat:
 
# Nous allons créer une fonction qui permet de déterminer automatiquement si l'employé ne fait plus partie des effectifs 
# en comparant la 'termination date' à la date du jour.
 
aujourdhui = pd.Timestamp.now().normalize()
def determine_active_status(date):
    if pd.to_datetime(date) < aujourdhui:
        return 0
    else:
        return 1
 
# Ensuite on applique cette fonction à la colonne 'active_status':
 
 
df_companydata_clean['Active_Status_corrige'] = df_companydata_clean['Termination_Date'].apply(lambda x: determine_active_status(x))
df_companydata_clean
```

```python
# La colonne active status n'a plus lieu d'être car elle est corrigée par la colonne activze status corrige, par conséquent je la supprime
df_companydata_clean = df_companydata_clean.drop('Active Status', axis=1)
```
